```python
import math
import pandas as pd
import seaborn as sns
import torch
from torch.distributions import transforms
import matplotlib.pyplot as plt
import pyro
import pyro.distributions as dist
from pyro.infer.mcmc import MCMC, NUTS
```

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
DATA_URL = "https://d2hg8soec8ck9v.cloudfront.net/datasets/rugged_data.csv"
rugged_data = pd.read_csv(DATA_URL, encoding="ISO-8859-1")
```

```python
rugged_data['log_gdp'] = rugged_data['rgdppc_2000'].apply(math.log)
rugged_data = rugged_data[rugged_data["rgdppc_2000"].notnull()]
df = rugged_data[['rugged','cont_africa','log_gdp']]
```

```python
df_rugged = torch.tensor(df["rugged"].values, dtype=torch.float)
df_cont_africa = torch.tensor(df["cont_africa"].values, dtype=torch.float)
df_log_gdp = torch.tensor(df["log_gdp"].values, dtype=torch.float)
```

```python
def model(rugged, cont_africa, log_gdp):
    a = pyro.sample("a", dist.Normal(0, 100)) # parameter의 prior 정의  N(0,100)
    bR = pyro.sample("bR", dist.Normal(0, 10))
    bA = pyro.sample("bA", dist.Normal(0, 10))
    bAR = pyro.sample("bAR", dist.Normal(0, 10))
    mu = a + bR * rugged + bA * cont_africa + bAR * rugged * cont_africa # rugged * cont_africa 는 interaction 반영
    sigma = pyro.sample("sigma", dist.Uniform(0, 10))
    with pyro.plate("plate"):
        pyro.sample("log_gdp", dist.Normal(mu, sigma), obs=log_gdp) # log_gdp = N(mu, sigma) 인 정규분포 가정 
```

```python
kernel = NUTS(model) # no u-turn sampling 
mcmc = MCMC(kernel, num_samples=1000)
mcmc.run(df_rugged, df_cont_africa, df_log_gdp) # hmc 를 사용 
```

```python
def summary(samples):
    site_stats = {}
    for site_name, values in samples.items():
        marginal_site = pd.DataFrame(values)
        describe = marginal_site.describe(percentiles=[.05, 0.25, 0.5, 0.75, 0.95]).transpose()
        site_stats[site_name] = describe[["mean", "std", "5%", "25%", "50%", "75%", "95%"]]
    return site_stats
hmc_samples = {k: v.detach().cpu().numpy() for k, v in mcmc.get_samples().items()}
for site, values in summary(hmc_samples).items():
    print("Site: {}".format(site))
    print(values, "\n")
```

```python
sites = ["a", "bA", "bR", "bAR", "sigma"]
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
for i, ax in enumerate(axs.reshape(-1)):
    site = sites[i]
    sns.distplot(hmc_samples[site], ax=ax, label="HMC")
    ax.set_title(site)

```
